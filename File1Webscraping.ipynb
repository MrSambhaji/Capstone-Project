```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException    

#Importing requests
import requests

# importing regex
import re
from selenium.webdriver.common.by import By
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
# Activating the chrome browser
#driver=webdriver.Chrome(r"chromedriver.exe") 
time.sleep(5)

# Opening the homepage of Amazon.in
url = "https://www.fabindia.com/clothing/men-western-wear"
driver.get(url)

time.sleep(5)
```

```python
#scrape all links for products
links=[]
for x in range(0,1):
    link=driver.find_elements(By.XPATH,'//a[@class="text-decoration-none text-muted text-secondary"]')
    for i in link:
        links.append(i.get_attribute("href"))
    
    time.sleep(3)
print(len(links))

#scrape require data from webpages
names=[]
Product_code=[]
sizes=[]
colors=[]
material_type=[]
Ocasions=[]
prices=[]
returns=[]
expected_delivery=[]
contry_of_origin=[]

for i in links:
    try:
        driver.get(i)
        try:
            name=driver.find_element(By.XPATH,'//h1[@class="prod_heading_title"]')
            names.append(name.text)
        except NoSuchElementException:
            names.append('-')
        
        try:
            Productcode=driver.find_element(By.XPATH,'//p[@class="pdp-sku"]')
            Product_code.append(Productcode.text)
        except NoSuchElementException:
            Product_code.append('-')
            
        try:
            size=driver.find_element(By.XPATH,'/html/body/app-root/cx-storefront/main/cx-page-layout/cx-page-slot[1]/app-fab-product-variants-container/div/app-fab-variant-size-selector/div/div[3]/div[1]/button')
            sizes.append(size.text)
        except NoSuchElementException:
            sizes.append('-')
            
        try:
            color=driver.find_element(By.XPATH,'/html/body/app-root/cx-storefront/main/cx-page-layout/cx-page-slot[1]/app-fab-product-variants-container/div/app-fab-variant-size-selector/div/div[1]/div/div/div/div[2]/img')
            colors.append(color.text)
        except NoSuchElementException:
            colors.append('-')
            
        try:
            material=driver.find_element(By.XPATH,'/html/body/app-root/cx-storefront/main/cx-page-layout/cx-page-slot[1]/fab-tab-paragraph-container/div/div/div[2]/div/div[1]/div/div/div/cx-product-attributes/div/div/table/tr[2]/td[3]/ul/li')
            material_type.append(material.text)
        except NoSuchElementException:
            material_type.append('-')
        
        try:
            Ocasion=driver.find_element(By.XPATH,'/html/body/app-root/cx-storefront/main/cx-page-layout/cx-page-slot[1]/fab-tab-paragraph-container/div/div/div[2]/div/div[1]/div/div/div/cx-product-attributes/div/div/table/tr[8]/td[3]/ul/li')
            Ocasions.append(Ocasion.text)
        except NoSuchElementException:
            Ocasions.append('-')

        try:
            price=driver.find_element(By.XPATH,'//div[@class="amount"]')
            prices.append(price.text.replace('\n','.'))
        except NoSuchElementException:
            prices.append('-')

        try:
            retur=driver.find_element(By.XPATH,'//h4[@class="pincode-serviceabilityTitle"]')
            returns.append(retur.text)
        except NoSuchElementException:
            returns.append('-')

        try:
            expected_D=driver.find_element(By.XPATH,'/html/body/app-root/cx-storefront/main/cx-page-layout/cx-page-slot[1]/fab-tab-paragraph-container/div/div/div[2]/div/div[1]/div/div/div/cx-product-attributes/div/div/table/tr[7]/td[3]/ul/li')
            expected_delivery.append(expected_D.text)
        except NoSuchElementException:
            expected_delivery.append('-')

        try:
            country_origin=driver.find_element(By.XPATH,'/html/body/app-root/cx-storefront/main/cx-page-layout/cx-page-slot[1]/fab-tab-paragraph-container/div/div/div[2]/div/div[1]/div/div/div/cx-product-attributes/div/div/table/tr[8]/td[3]/ul/li')
            contry_of_origin.append(country_origin.text)
        except NoSuchElementException:
            contry_of_origin.append('-')
    except TimeoutError:
        driver.wait(10)
        names.append('-')
        colors.append('-')
        sizes.append('-')
        Product_code.append('-')
        material_type.append('-')
        Ocasions.append('-')
        prices.append('-')
        returns.append('-')
        expected_delivery.append('-')
        contry_of_origin.append('-')
    time.sleep(10)
    
#creating dataframe
df=pd.DataFrame({
    'name':names,
    'product code':Product_code,
    'product size':sizes,
    'color available':colors,
    'materail type':material_type,
    'ocasions':Ocasions,
    'price':prices,
    'returns':returns,
    'expected_delivery':expected_delivery,
    'country of origin':contry_of_origin
})
df
```

```python
#Importing Libraries
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
url="https://www.ajio.com/men-sweatshirt-hoodies/c/830216011"
```

```python
driver.get(url)
```

```python
last_height=0
```

```python
#list to store scrapped data
brand=[]
product_name=[]
rating=[]
num_rating=[]
price=[]
original_prices=[]
discounts=[]
offer_price=[]
```

```python
#To scroll down for loading content
SCROLL_PAUSE_TIME=20
```

```python
while True:
    #Get dynamically loaded HTML
    html=driver.page_source
    soup=BeautifulSoup(html, 'html.parser')
    #Extracting the data
    items=soup.find_all(class_='item')
    
    for item in items[:100]:
        brand_elem=item.find(class_='brand')
        product_name_elem=item.find(class_='nameCls')
        rating_elem=item.find(class_='_1N0OO')
        parent_div=item.find(class_='_2QgMK _13cxi _1BsqP _2564l')
        num_rating_elem=parent_div.find_all('p')[1].text.strip() if parent_div else 'N/A'
        price_elem=item.find(class_='price')
        original_price_elem=item.find(class_='original-price')
        discount_elem=item.find(class_='discount')
        offer_price_elem=item.find(class_='offer-pricess')
        
        #Append data to lits
        brand.append(brand_elem.text if brand_elem else "N/A")
        product_name.append(product_name_elem.text if product_name_elem else "N/A")
        rating.append(rating_elem.text if rating_elem else "N/A")
        num_rating.append(num_rating_elem)
        price.append(price_elem.text if price_elem else "N/A")
        original_prices.append(original_price_elem.text if original_price_elem else "N/A")
        discounts.append(discount_elem.text if discount_elem else "N/A")
        offer_price.append(offer_price_elem.text if offer_price_elem else "N/A")
        
     #Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
  
     #Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    
     #Calculate new scroll height and compare with last scroll height
    new_height=driver.execute_script("return document.body.scrollHeight")
    if new_height==last_height:
        #If no more content is loaded, exit the loop
        break
    last_height=new_height
    
    
#Close the webdriver
driver.quit()
```

```python
#Crate DataFrame
df=pd.DataFrame({
    'Brand': brand,
    'Product Name': product_name,
    'Rating': rating,
    'Number of Ratings': num_rating,
    'price': price,
    'Original Price': original_prices,
    'Discount': discounts,
    'Offer Price': offer_price
})

print(df)
```
